### Q2: What are the three different types of Boolean operators?

**Answer:** In Python, there are three main Boolean operators:

AND (and): This operator returns True if both operands are True, otherwise, it returns False.

OR (or): The or operator returns True if at least one of the operands is True. If both operands are False, it returns False.

NOT (not): This operator negates the value of the operand. If the operand is True, not will return False, and if the operand is False, not will return True.

### Q3: Make a list of each Boolean operator's truth tables (i.e. every possible combination of Boolean  values for the operator and what it evaluate).

**Answer:** The truth tables for the three basic Boolean operators are as follows:

1. **AND Operator (&& or AND):**

| A     | B     | A AND B |
|-------|-------|---------|
| False | False | False   |
| False | True  | False   |
| True  | False | False   |
| True  | True  | True    |

2. **OR Operator (|| or OR):**

| A     | B     | A OR B  |
|-------|-------|---------|
| False | False | False   |
| False | True  | True    |
| True  | False | True    |
| True  | True  | True    |

3. **NOT Operator (! or NOT):**

| A     | NOT A |
|-------|-------|
| False | True  |
| True  | False |

- `==` is a comparison operator used to check if two values are equal. For example, `x == y` checks if the value of `x` is equal to the value of `y`.

- `=` is an assignment operator used to assign a value to a variable. For example, `x = 5` assigns the value `5` to the variable `x`.

We would use `==` in conditions to compare values, like in an `if` statement to check if two values are equal. We would use `=` when we want to store or update a value in a variable. For example:

```python
x = 5             # Assigns the value 5 to variable x
if x == 10:       # Checks if x is equal to 10
    # Code to execute if x is equal to 10
```

In the example above, `=` is used for assignment, and `==` is used for comparison within the 'if' statement.

### Q10: How can you tell the difference between break and continue?

**Answer:** In a very short way:

- `break` is used to exit a loop prematurely, terminating the loop entirely.
- `continue` is used to skip the current iteration of a loop and proceed to the next iteration.

### Q11: In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

**Answer:** In a `for` loop in Python, there is no practical difference between `range(10)`, `range(0, 10)`, and `range(0, 10, 1)`:

- `range(10)`: It generates numbers from 0 to 9 (inclusive), incrementing by the default step of 1.

- `range(0, 10)`: It's equivalent to `range(10)` and generates numbers from 0 to 9 (inclusive), incrementing by the default step of 1.

- `range(0, 10, 1)`: It's also equivalent to `range(10)` and generates numbers from 0 to 9 (inclusive), incrementing by 1.

All three expressions will create the same sequence of numbers, and we can use any of them interchangeably in a `for` loop.

### Q12: Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent program that prints the numbers 1 to 10 using a while loop.

**Answer:**

```python
# Using a for loop:

for i in range(1, 11):
    print(i)
  
    
# Using a while loop:   
i = 1
while i <= 10:
    print(i)
    i += 1

```
